```python
# Import libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Ignore warning messages
import warnings
warnings.filterwarnings("ignore")
```

```python
# Load the dataset

df = pd.read_csv("creditcard.csv")
```

```python
df.head()
```

```python
# Number of rows and columns

print("Dataset Shape:", df.shape)
```

```python
# Column names

print(df.columns)
```

```python
# Dataset information

df.info()
```

```python
# Check missing values

df.isnull().sum()
```

```python
# Statistical summary

df.describe()
```

```python
# Count the number of genuine and fraud transactions

df["Class"].value_counts()
```

```python
# Calculate fraud percentage

fraud = df[df["Class"] == 1].shape[0]
genuine = df[df["Class"] == 0].shape[0]
total = len(df)

print("Total Transactions :", total)
print("Genuine Transactions :", genuine)
print("Fraud Transactions :", fraud)

print("\nFraud Percentage :", round((fraud/total)*100,4), "%")
```

```python
# Visualize class distribution

df["Class"].value_counts().plot(kind="bar")

plt.title("Fraud vs Genuine Transactions")
plt.xlabel("Class")
plt.ylabel("Number of Transactions")

plt.show()
```

```python
# Pie chart

df["Class"].value_counts().plot(
    kind="pie",
    autopct="%1.2f%%",
    labels=["Genuine", "Fraud"]
)

plt.title("Transaction Distribution")
plt.ylabel("")

plt.show()
```

```python
# Features and Target

X = df.drop("Class", axis=1)
y = df["Class"]

print("Features Shape :", X.shape)
print("Target Shape :", y.shape)
```

```python
from sklearn.ensemble import IsolationForest
```

```python
# Create the Isolation Forest model

model = IsolationForest(
    n_estimators=100,
    contamination=0.002,
    random_state=42
)

# Train the model
model.fit(X)

print("Model trained successfully!")
```

```python
# Predict transactions

predictions = model.predict(X)

print(predictions[:20])
```

```python
# Convert predictions to match dataset labels

predictions = np.where(predictions == -1, 1, 0)

print(predictions[:20])
```

```python
from sklearn.metrics import (
    accuracy_score,
    precision_score,
    recall_score,
    f1_score,
    confusion_matrix,
    classification_report
)
```

```python
# Accuracy

accuracy = accuracy_score(y, predictions)

print("Accuracy :", accuracy)
```

```python
# Precision

precision = precision_score(y, predictions)

print("Precision :", precision)
```

```python
# Recall

recall = recall_score(y, predictions)

print("Recall :", recall)
```

```python
# F1 Score

f1 = f1_score(y, predictions)

print("F1 Score :", f1)
```

```python
# Confusion Matrix

cm = confusion_matrix(y, predictions)

print(cm)
```

```python
# Complete Report

print(classification_report(y, predictions))
```

```python
import joblib

# Save the trained model
joblib.dump(model, "model.pkl")

print("Model saved successfully!")
```
